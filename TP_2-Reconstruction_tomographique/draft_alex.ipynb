### Contexte

<div style="background-color: #252526; color: white; border-left: 5px solid #9C27B0; padding: 15px; border-radius: 5px; margin-top: 10px;">

<strong style="color: #CE93D8; font-size: 1.05em;">Tomographie :</strong><br><br>
La **tomodensitom√©trie** (TDM / CT scan) est une technique d'imagerie non invasive permettant de reconstruire la distribution de densit√© d'un objet tranche par tranche. Le principe repose sur la mesure de l'**att√©nuation partielle** d'un faisceau de rayons X traversant l'objet sous diff√©rents angles.

Les fondements th√©oriques ont √©t√© √©nonc√©s par le math√©maticien **Johann Radon** en 1917. L'invention du tomodensitom√®tre m√©dical est attribu√©e √† **Sir Godfrey Hounsfield** et **Allan McLeod Cormack** (prix Nobel de m√©decine 1979).

**G√©om√©trie :** On utilise ici une g√©om√©trie √† **faisceaux parall√®les**, o√π le param√®tre $t \in [0, 1]$ d√©signe la position relative sur le d√©tecteur et $\varphi \in [0, 2\pi[$ l'angle du statif. L'ensemble des mesures forme le **sinogramme**.

</div>

### Loi de Beer-Lambert

L'att√©nuation des rayons X est mod√©lis√©e par :

$$I = I_0 \exp\left(-\int \mu(x)\,dx\right) \quad \Longrightarrow \quad \ln\frac{I_0}{I} = \sum \mu(x)\,\Delta x$$

o√π $\mu(x)$ est le **coefficient d'att√©nuation lin√©aire** (la quantit√© √† reconstruire).

### R√©f√©rence

> M. Slaney and A. Kak, *Principles of Computerized Tomographic Imaging*, IEEE Press, 1988.

**√ânonc√© :** *Supposons, pour simplifier, que le corps humain est fait d'eau pure. Gr√¢ce √† la base de donn√©es XCOM du NIST ([lien](http://physics.nist.gov/PhysRefData/Xcom/html/xcom1.html)), trouvez le ¬µ de l'eau pour des photons de 60 keV, √† une masse volumique de 1.0 g/cm¬≥. Quelle fraction du signal subsiste apr√®s avoir travers√© 5 cm d'eau ; 20 cm d'eau ?*

</div>

```python
# Donn√©es NIST XCOM pour l'eau √† 60 keV
# https://physics.nist.gov/PhysRefData/Xcom/html/xcom1.html
# Pour H2O √† 60 keV : Œº/œÅ = 0.2059 cm¬≤/g

mu_sur_rho = 0.2059  # cm¬≤/g (coefficient d'att√©nuation massique)
rho_eau = 1.0        # g/cm¬≥ (masse volumique de l'eau)

# Coefficient d'att√©nuation lin√©aire
mu = mu_sur_rho * rho_eau  # cm‚Åª¬π

# Loi de Beer-Lambert : I = I‚ÇÄ exp(-Œºx)
# Cas 1 : 5 cm d'eau
x1 = 5.0  # cm
fraction_5cm = np.exp(-mu * x1)
attenuation_5cm = (1 - fraction_5cm) * 100

# Cas 2 : 20 cm d'eau
x2 = 20.0  # cm
fraction_20cm = np.exp(-mu * x2)
attenuation_20cm = (1 - fraction_20cm) * 100

print(f"Œº = {mu:.4f} cm‚Åª¬π")
print(f"Travers√©e 5 cm  : {fraction_5cm:.4f} ({fraction_5cm*100:.2f}% transmis)")
print(f"Travers√©e 20 cm : {fraction_20cm:.6f} ({fraction_20cm*100:.4f}% transmis)")
```

### Interpr√©tation physique

#### D√©croissance exponentielle

Les graphiques montrent clairement la **nature exponentielle** de l'att√©nuation :
- **√âchelle lin√©aire** : d√©croissance rapide dans les premiers centim√®tres
- **√âchelle logarithmique** : relation **lin√©aire** ‚Üí confirme $\ln(I/I_0) = -\mu x$

#### Implications pour la tomographie

En tomodensitom√©trie (TDM), on mesure l'intensit√© transmise $I$ pour retrouver le coefficient d'att√©nuation :

$$-\ln\left(\frac{I}{I_0}\right) = \mu \cdot x$$

Chaque **projection** du sinogramme correspond √† l'int√©grale des coefficients d'att√©nuation le long d'un rayon. La reconstruction consiste √† **inverser** cette transformation de Radon pour retrouver la distribution spatiale de $\mu(x, y)$.

**Points cl√©s :**
- L'att√©nuation augmente exponentiellement avec l'√©paisseur
- Le logarithme transforme la relation en somme lin√©aire
- Cette propri√©t√© est fondamentale pour la reconstruction tomographique

```python
# Visualisation : Courbe d'att√©nuation en fonction de l'√©paisseur
x_range = np.linspace(0, 30, 500)
I_sur_I0 = np.exp(-mu * x_range)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))

# Graphique 1 : √âchelle lin√©aire
ax1.plot(x_range, I_sur_I0, 'b-', linewidth=2.5, label=r'$I/I_0 = e^{-\mu x}$')
ax1.axvline(5, color='red', linestyle='--', alpha=0.7, linewidth=2, label=f'x = 5 cm ‚Üí {fraction_5cm:.1%}')
ax1.axvline(20, color='orange', linestyle='--', alpha=0.7, linewidth=2, label=f'x = 20 cm ‚Üí {fraction_20cm:.2%}')
ax1.scatter([5, 20], [fraction_5cm, fraction_20cm], color=['red', 'orange'], s=150, zorder=5, edgecolors='black', linewidths=1.5)
ax1.set_xlabel('√âpaisseur d\'eau travers√©e (cm)', fontsize=13, fontweight='bold')
ax1.set_ylabel(r'Fraction transmise $I/I_0$', fontsize=13, fontweight='bold')
ax1.set_title('Loi de Beer-Lambert ‚Äî √âchelle lin√©aire', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=11, loc='upper right')
ax1.set_xlim(0, 30)
ax1.set_ylim(0, 1.05)

# Graphique 2 : √âchelle logarithmique
ax2.semilogy(x_range, I_sur_I0, 'b-', linewidth=2.5, label=r'$I/I_0 = e^{-\mu x}$')
ax2.axvline(5, color='red', linestyle='--', alpha=0.7, linewidth=2, label=f'x = 5 cm ‚Üí {fraction_5cm:.1%}')
ax2.axvline(20, color='orange', linestyle='--', alpha=0.7, linewidth=2, label=f'x = 20 cm ‚Üí {fraction_20cm:.2%}')
ax2.scatter([5, 20], [fraction_5cm, fraction_20cm], color=['red', 'orange'], s=150, zorder=5, edgecolors='black', linewidths=1.5)
ax2.set_xlabel('√âpaisseur d\'eau travers√©e (cm)', fontsize=13, fontweight='bold')
ax2.set_ylabel(r'Fraction transmise $I/I_0$ (log)', fontsize=13, fontweight='bold')
ax2.set_title('Loi de Beer-Lambert ‚Äî √âchelle logarithmique', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, which='both')
ax2.legend(fontsize=11, loc='upper right')
ax2.set_xlim(0, 30)

fig.suptitle('Att√©nuation des rayons X de 60 keV dans l\'eau', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## Approche ‚Äî R√©troprojection voxel-driven

L'algorithme de **r√©troprojection simple** (ou laminogramme) est la m√©thode la plus intuitive pour reconstruire une image CT. Pour chaque voxel $(i, j)$ de la grille de reconstruction, on calcule sa projection sur le d√©tecteur pour chaque angle $\theta$ et on somme les valeurs mesur√©es :

$$g(x, y) = \sum_{\theta} p_\theta(x \cos\theta + y \sin\theta)$$

L'identification du pixel d√©tecteur se fait par **plus proche voisin** : $\text{idx} = \text{round}(\text{center}_{\text{det}} + t / p_{\text{size}})$.

**Complexit√© algorithmique :** $\mathcal{O}(N_{\text{proj}} \times N_{\text{vox}}^2)$ ‚Äî triple boucle imbriqu√©e, tr√®s lent pour de grandes grilles.

**Tableau 3 : Caract√©ristiques du laminogramme**

| Aspect | Description |
|:-------|:------------|
| **M√©thode** | Voxel-driven (parcours de la grille de reconstruction) |
| **Interpolation** | Plus proche voisin sur le d√©tecteur |

| **Avantages** | Simple √† impl√©menter, intuitif || **Inconv√©nients** | Tr√®s lent, qualit√© d'image m√©diocre (pas de filtrage) |

```python
def laminogram_simple(sinogram, angles, nbvox=96, voxsize=0.4, pixsize=0.165):
    """
    R√©troprojection simple (laminogramme) - m√©thode voxel-driven

    Pour chaque voxel, projeter sur toutes les directions et sommer.
    """
    nbprj, nbpix = sinogram.shape
    image = np.zeros((nbvox, nbvox))

    center = (nbvox - 1) / 2.0
    det_center = (nbpix - 1) / 2.0

    # Boucle sur chaque voxel
    for j in range(nbvox):
        if j % 20 == 0:
            print(f"  Progression : {j}/{nbvox} lignes ({100*j/nbvox:.0f}%)", end='\r')
        for i in range(nbvox):
            # Position du voxel dans le rep√®re monde (cm)
            x = (i - center) * voxsize
            y = (j - center) * voxsize

            # Somme sur toutes les projections
            for ia, angle in enumerate(angles):
                # Projection sur le d√©tecteur : t = x¬∑cos(Œ∏) + y¬∑sin(Œ∏)
                t = x * np.cos(angle) + y * np.sin(angle)

                # Index du pixel d√©tecteur correspondant
                idx_det = det_center + t / pixsize

                # Plus proche voisin
                idx_det_int = int(np.round(idx_det))

                # V√©rifier que l'index est valide
                if 0 <= idx_det_int < nbpix:
                    image[j, i] += sinogram[ia, idx_det_int]

    print(f"  Progression : {nbvox}/{nbvox} lignes (100%)    ")
    return image

# Param√®tres de reconstruction
nbvox = 96         # Nombre de voxels par dimension
voxsize = 0.4      # Taille d'un voxel (cm)
pixsize = 0.165    # Taille d'un pixel du d√©tecteur (cm)

# Reconstruction par laminogramme
print("Reconstruction par laminogramme (r√©troprojection simple)...")
t_start = time.time()
image_lam = laminogram_simple(sinogram_password, angles, nbvox, voxsize, pixsize)
t_end = time.time()
temps_lam = t_end - t_start

print(f"Temps de calcul : {temps_lam:.2f} secondes")
```

### Rappel th√©orique ‚Äî Algorithme de filtrage

L'op√©ration de filtrage dans le domaine fr√©quentiel s'effectue en trois √©tapes :

$$P(u) = \mathcal{F}[p(t)] \qquad \longrightarrow \qquad P_f(u) = P(u) \times |u| \qquad \longrightarrow \qquad p_f(t) = \mathcal{F}^{-1}[P(u) \times |u|]$$

Le filtre $f(u) = |u|$ amplifie les hautes fr√©quences spatiales, ce qui compense l'exc√®s de basses fr√©quences introduit par la r√©troprojection.

<div style="background-color: #1a1a2e; color: #a8d0e6; border-left: 5px solid #FFB300; padding: 12px; border-radius: 5px; margin-top: 10px; font-size: 0.9em;">

üí° **Indices :** Lisez bien √† quoi sert la fonction `fftshift` et n'oubliez pas que $p(t)$ est une fonction r√©elle.

</div>

```python
def filter_projection_ramp(projection):
    """
    Applique le filtre passe-haut |u| (filtre rampe) √† une projection 1D
    
    √âtapes :
      1. Transform√©e de Fourier 1D : P(u) = FFT[p(t)]
      2. Application du filtre rampe : P_f(u) = P(u) √ó |u|
      3. Transform√©e inverse : p_f(t) = IFFT[P_f(u)]
    """
    N = len(projection)
    
    # 1. Transform√©e de Fourier 1D
    P = fft(projection)
    
    # 2. Cr√©ation du filtre rampe |u|
    freqs = fftfreq(N)  # Fr√©quences normalis√©es [-0.5, 0.5]
    ramp_filter = np.abs(freqs)
    
    # 3. Application du filtre dans le domaine fr√©quentiel
    P_filtered = P * ramp_filter
    
    # 4. Transform√©e inverse pour revenir dans le domaine spatial
    projection_filtered = ifft(P_filtered)
    
    # 5. Retourner la partie r√©elle (le signal est r√©el)
    return np.real(projection_filtered)


def filter_sinogram_ramp(sinogram):
    """
    Filtre toutes les projections d'un sinogramme avec le filtre rampe
    """
    nbprj, nbpix = sinogram.shape
    sinogram_filtered = np.zeros_like(sinogram)
    
    for i in range(nbprj):
        sinogram_filtered[i] = filter_projection_ramp(sinogram[i])
    
    return sinogram_filtered

# Application du filtre
print("Filtrage du sinogramme avec le filtre rampe |u|...")
t_start = time.time()
sinogram_filtered = filter_sinogram_ramp(sinogram_password)
t_end = time.time()
temps_filtre = t_end - t_start
print(f"Temps de filtrage : {temps_filtre:.3f} secondes")
```

## Principe du filtrage rampe

Le **filtre rampe** (ou filtre $|u|$) est un filtre passe-haut essentiel en tomographie. Il compense la d√©croissance en $1/r$ du laminogramme en appliquant √† chaque projection :

$$p_{\text{filtr√©}}(t) = \mathcal{F}^{-1}\left[\mathcal{F}[p(t)] \cdot |u|\right]$$

Pour chaque projection du sinogramme, on effectue :

1. **Transform√©e de Fourier** : $P(u) = \text{FFT}[p(t)]$
2. **Application du filtre** : $P_{\text{filtr√©}}(u) = P(u) \times |u|$
3. **Transform√©e inverse** : $p_{\text{filtr√©}}(t) = \text{Re}[\text{IFFT}[P_{\text{filtr√©}}(u)]]$

Le filtre $H(u) = |u|$ amplifie lin√©airement les hautes fr√©quences spatiales. Ses propri√©t√©s :

**Tableau 4 : Propri√©t√©s du filtre rampe**

| Aspect | Description |
|:-------|:------------|
| **Type** | Passe-haut |
| **$H(0) = 0$** | Supprime la composante continue (biais) |

| **$H(u) \propto u$** | Amplifie lin√©airement les hautes fr√©quences || **Co√ªt** | $\mathcal{O}(N_{\text{proj}} \times N_{\text{pix}} \log N_{\text{pix}})$ |

## Principe ‚Äî Interpolation lin√©aire

Au lieu du **plus proche voisin** (Q4), on interpole lin√©airement entre les deux pixels adjacents du d√©tecteur. Si la projection d'un voxel tombe √† l'indice $t = 42.7$ :

**Tableau 6 : Interpolation PPV vs lin√©aire**

| M√©thode | Valeur utilis√©e |
|:--------|:----------------|
| Plus proche voisin (Q4) | `sino[43]` |
| Interpolation lin√©aire (Q5) | $0.3 \times \texttt{sino[42]} + 0.7 \times \texttt{sino[43]}$ |

Cette approche r√©duit les artefacts de discr√©tisation et lisse les transitions abruptes entre pixels voisins.

```python
# Question 5 : FBP avec interpolation lin√©aire

def backproject_linear_interp(sinogram, angles, nbvox=96, voxsize=0.4, pixsize=0.165):
    """
    R√©troprojection filtr√©e avec interpolation lin√©aire (1D).

    Au lieu du plus proche voisin (round), on interpole lin√©airement
    entre les deux pixels adjacents du d√©tecteur :
      t_pixel = 42.7 ‚Üí val = 0.3 * sino[42] + 0.7 * sino[43]
    """
    nbprj, nbpix = sinogram.shape
    image = np.zeros((nbvox, nbvox))
    center = (nbvox - 1) / 2.0
    det_center = (nbpix - 1) / 2.0

    for j in range(nbvox):
        if j % 20 == 0:
            print(f"  Progression : {j}/{nbvox} lignes ({100*j/nbvox:.0f}%)", end='\r')
        for i in range(nbvox):
            x = (i - center) * voxsize
            y = (j - center) * voxsize

            for a_idx, phi in enumerate(angles):
                t = x * np.cos(phi) + y * np.sin(phi)
                t_pixel = t / pixsize + det_center

                idx_left = int(np.floor(t_pixel))
                idx_right = idx_left + 1
                frac = t_pixel - idx_left

                if 0 <= idx_left < nbpix and 0 <= idx_right < nbpix:
                    value = (1 - frac) * sinogram[a_idx, idx_left] + \
                            frac * sinogram[a_idx, idx_right]
                    image[j, i] += value
                elif 0 <= idx_left < nbpix:
                    image[j, i] += sinogram[a_idx, idx_left]
                elif 0 <= idx_right < nbpix:
                    image[j, i] += sinogram[a_idx, idx_right]

    print(f"  Progression : {nbvox}/{nbvox} lignes (100%)    ")
    return image


# Reconstruction FBP avec interpolation lin√©aire
t_start = time.time()
image_fbp_linear = backproject_linear_interp(sinogram_filtered, angles, nbvox, voxsize, pixsize)
t_end = time.time()
temps_fbp_linear = t_end - t_start
temps_fbp_q5 = temps_fbp_linear

print(f"Temps de reconstruction : {temps_fbp_linear:.2f} s")

# --- Visualisation : PPV vs interpolation lin√©aire ---
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

vmin1, vmax1 = np.percentile(image_fbp, [1, 99])
im0 = axes[0].imshow(image_fbp, cmap='gray', interpolation='nearest', vmin=vmin1, vmax=vmax1)
axes[0].set_title(f'FBP ‚Äî Plus proche voisin\n(Q4, temps: {temps_fbp:.1f}s)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('x (pixel)')
axes[0].set_ylabel('y (pixel)')
divider0 = make_axes_locatable(axes[0])
cax0 = divider0.append_axes("right", size="5%", pad=0.1)
plt.colorbar(im0, cax=cax0, label=r'$\mu$ (cm$^{-1}$)')

vmin2, vmax2 = np.percentile(image_fbp_linear, [1, 99])
im1 = axes[1].imshow(image_fbp_linear, cmap='gray', interpolation='nearest', vmin=vmin2, vmax=vmax2)
axes[1].set_title(f'FBP ‚Äî Interpolation lin√©aire\n(Q5, temps: {temps_fbp_linear:.1f}s)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('x (pixel)')
axes[1].set_ylabel('y (pixel)')
divider1 = make_axes_locatable(axes[1])
cax1 = divider1.append_axes("right", size="5%", pad=0.1)
plt.colorbar(im1, cax=cax1, label=r'$\mu$ (cm$^{-1}$)')

diff = image_fbp_linear - image_fbp
im2 = axes[2].imshow(diff, cmap='RdBu_r', interpolation='nearest')
axes[2].set_title('Diff√©rence\n(lin√©aire - plus proche voisin)', fontsize=13, fontweight='bold')
axes[2].set_xlabel('x (pixel)')
axes[2].set_ylabel('y (pixel)')
divider2 = make_axes_locatable(axes[2])
cax2 = divider2.append_axes("right", size="5%", pad=0.1)
plt.colorbar(im2, cax=cax2, label=r'$\Delta\mu$ (cm$^{-1}$)')

fig.suptitle('Question 5 ‚Äî Impact de l\'interpolation lin√©aire sur la FBP',
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

# --- Zoom ROI central ---
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

roi_size = 30
center_idx = nbvox // 2
roi_slice = (slice(center_idx - roi_size//2, center_idx + roi_size//2),
             slice(center_idx - roi_size//2, center_idx + roi_size//2))

roi_ppv = image_fbp[roi_slice]
roi_linear = image_fbp_linear[roi_slice]

im0 = axes[0].imshow(roi_ppv, cmap='gray', interpolation='nearest')
axes[0].set_title('ROI ‚Äî Plus proche voisin', fontsize=13, fontweight='bold')
plt.colorbar(im0, ax=axes[0], label=r'$\mu$ (cm$^{-1}$)')

im1 = axes[1].imshow(roi_linear, cmap='gray', interpolation='nearest')
axes[1].set_title('ROI ‚Äî Interpolation lin√©aire', fontsize=13, fontweight='bold')
plt.colorbar(im1, ax=axes[1], label=r'$\mu$ (cm$^{-1}$)')

fig.suptitle('Zoom sur r√©gion centrale (30√ó30 pixels)', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Rappel th√©orique ‚Äî Fourier Slice Theorem

La transform√©e de Fourier d'une projection 1D √† un angle $\varphi$ correspond √† une ligne de la TF 2D de l'image recherch√©e, au m√™me angle :

$$\mathcal{F}[p(t)]_{\varphi = \text{const}} = \mathrm{TF}_{2D}[\mu(t, \varphi = \text{const})]$$

On peuple donc l'espace de Fourier $M(u, v)$ avec les lignes de projection $P(t)_\varphi$, puis on applique $\mathrm{TF}^{-1}_{2D}$ pour obtenir l'image.

<div style="background-color: #1a1a2e; color: #a8d0e6; border-left: 5px solid #FFB300; padding: 12px; border-radius: 5px; margin-top: 10px; font-size: 0.9em;">

üí° **Indice :** Initialisez la matrice Fourier 2D en complexe : `mymatrix = numpy.zeros((a, b), 'complex')` afin d'√©viter les changements de taille, puis prenez la partie r√©elle de l'image finale.

</div>

```python
def recon_fourier_slice(sinogram, angles, nbvox, voxsize, pixsize):
    """
    Reconstruction par th√©or√®me de la tranche de Fourier (Fourier Slice Theorem).

    Points critiques :
    1) Le sinogramme est centr√© (s=0 au centre du d√©tecteur) -> ifftshift AVANT FFT 1D
    2) Mapping radial via fftfreq(..., d=pixsize) et fftfreq(..., d=voxsize)
    3) Recentrer l'image reconstruite -> fftshift apr√®s iFFT 2D
    """
    nbprj, nbpix = sinogram.shape

    # FFT des projections (centrage indispensable)
    SINOGRAM = fftshift(fft(ifftshift(sinogram, axes=1), axis=1), axes=1)

    # Axes de fr√©quences (cycles / cm)
    freq_detector = fftshift(fftfreq(nbpix, d=pixsize))
    freq_img      = fftshift(fftfreq(nbvox, d=voxsize))

    IMAGE_FOURIER = np.zeros((nbvox, nbvox), dtype=complex)

    # Remplissage du plan de Fourier (interpolation radiale + angle le plus proche)
    for v_idx, v in enumerate(freq_img):
        for u_idx, u in enumerate(freq_img):
            rho = np.sqrt(u*u + v*v)
            phi = np.arctan2(v, u)
            if phi < 0:
                phi += 2*np.pi

            # distance angulaire circulaire (0..pi)
            angle_diff = np.abs(((angles - phi + np.pi) % (2*np.pi)) - np.pi)
            angle_idx = int(np.argmin(angle_diff))

            # interpolation radiale sur la FFT 1D
            rho_idx = np.interp(rho, freq_detector, np.arange(nbpix), left=np.nan, right=np.nan)
            if np.isnan(rho_idx):
                continue

            i0 = int(np.floor(rho_idx))
            i1 = min(i0 + 1, nbpix - 1)
            frac = rho_idx - i0

            IMAGE_FOURIER[v_idx, u_idx] = (1 - frac) * SINOGRAM[angle_idx, i0] + frac * SINOGRAM[angle_idx, i1]

    # iFFT 2D + recentrage pour affichage
    image = fftshift(ifft2(ifftshift(IMAGE_FOURIER))).real
    return image

# Application de la reconstruction via Fourier Slice Theorem
print("Reconstruction via Fourier Slice Theorem...")
t_start = time.time()
image_fourier = recon_fourier_slice(sinogram_password, angles, nbvox, voxsize, pixsize)
t_end = time.time()
temps_fourier = t_end - t_start

print(f"Temps de reconstruction : {temps_fourier:.2f} secondes")
```

## Observations pr√©liminaires

### Analyse comparative des performances

**Tableau 11 : Complexit√© algorithmique des m√©thodes de reconstruction**

| M√©thode | Complexit√© (impl√©mentation) | Commentaire |
|:--------|:---------------------------|:------------|
| **Laminogramme** | $\mathcal{O}(N_{\text{proj}} \times N_{\text{vox}}^2)$ | Acc√®s O(1) au d√©tecteur (round + index) |
| **FBP (PPV / lin√©aire)** | $\mathcal{O}(N_{\text{proj}} \times N_{\text{pix}} + N_{\text{proj}} \times N_{\text{vox}}^2)$ | Filtrage O(N_proj √ó N_pix) + r√©troprojection O(N_proj √ó N_vox¬≤) |
| **Fourier Slice 96√ó96** | $\mathcal{O}(N_{\text{proj}} \times N_{\text{pix}} \log N_{\text{pix}} + N_{\text{vox}}^2 \times N_{\text{proj}})$ | FFT 1D + remplissage 2D avec argmin sur angles |
| **Fourier Slice 192√ó192** | Idem avec $N_{\text{vox}} = 192$ | Temps √ó4 environ (quadratique en N_vox) |

#### Pourquoi le Fourier Slice n'est pas forc√©ment plus rapide que la FBP

Dans notre impl√©mentation, le Fourier Slice Theorem n'est pas n√©cessairement plus rapide que la FBP car la boucle de remplissage 2D contient une **recherche d'angle par `argmin`** √† chaque pixel de la grille : `angle_idx = np.argmin(np.abs(angles - phi))`, ce qui est $\mathcal{O}(N_{\text{proj}})$ par pixel. La complexit√© th√©orique optimale (avec gridding pr√©-calcul√© et interpolation directe) serait $\mathcal{O}(N_{\text{proj}} \times N_{\text{pix}} \log N_{\text{pix}} + N_{\text{vox}}^2 \log N_{\text{vox}}^2)$.

#### Discussion sur les boucles `for` imbriqu√©es et l'utilit√© du slicing

- **Laminogramme (Q2)** : 3 boucles imbriqu√©es (j, i, angles) ‚Äî la boucle interne sur les angles parcourt toutes les projections pour chaque voxel.
- **FBP (Q4-Q5)** : m√™me structure √† 3 boucles, mais le filtrage pr√©alable est effectu√© par **slicing numpy** (op√©rations vectoris√©es sur les pixels du d√©tecteur via FFT/IFFT), ce qui est beaucoup plus efficace qu'une boucle Python.
- **Fourier Slice (Q6)** : 2 boucles imbriqu√©es (v_idx, u_idx) sur la grille de Fourier 2D, mais avec un `argmin` co√ªteux √† l'int√©rieur ($\mathcal{O}(N_{\text{proj}})$).

Le **slicing numpy** (op√©rations vectoris√©es) pourrait remplacer certaines boucles `for` pour acc√©l√©rer significativement le code. Par exemple, le filtrage du sinogramme (Q3) utilise d√©j√† `fft`/`ifft` vectoris√©s. La r√©troprojection pourrait √©galement √™tre vectoris√©e en calculant les projections de tous les voxels d'une ligne simultan√©ment (slicing sur l'axe `i`), √©liminant la boucle interne.

#### Points cl√©s

**Laminogramme (Q2) :**
Triple boucle imbriqu√©e (voxel-driven) sans optimisation ‚Üí **tr√®s lent** et qualit√© insuffisante.

**FBP + interpolation lin√©aire (Q5) :**
Meilleur **compromis qualit√©/temps**, surco√ªt minime par rapport √† FBP basique (Q4). **Gold standard** clinique.

**Fourier Slice 96√ó96 (Q6) :**
FFT tr√®s efficace, mais le remplissage 2D avec `argmin` limite les gains. Qualit√© d√©grad√©e (artefacts Nyquist).

**Fourier Slice 192√ó192 (Q8) :**
Qualit√© **proche de FBP** gr√¢ce au sur-√©chantillonnage. Temps √ó4 par rapport √† Q6 (quadratique en $N_{\text{vox}}$).

## Principe ‚Äî Interpolation avec lissage angulaire et interpolation lin√©aire radiale

Au lieu d'utiliser **uniquement le plus proche voisin** en angle (Q6), cette approche combine :
- **Interpolation lin√©aire radiale** : comme en Q6, on interpole entre les 2 fr√©quences radiales voisines
- **Lissage angulaire** : on pond√®re les contributions des projections voisines (¬±1 angle) par leur proximit√© angulaire

Cette approche interpole les valeurs Fourier en utilisant les projections voisines (¬±1 angle) pond√©r√©es par la proximit√© angulaire, combin√©e √† une interpolation lin√©aire sur les fr√©quences radiales. **Ce n'est pas une interpolation bilin√©aire au sens strict** (4 voisins sur une grille r√©guli√®re), mais plut√¥t une approximation adapt√©e aux coordonn√©es polaires de l'espace de Fourier.

**Tableau 12 : Comparaison Q6 vs Q10**

| Aspect | Q6 (Nearest Neighbor) | Q10 (Lissage angulaire + interp. radiale) |
|:-------|:---------------------|:-----------------|
| Interpolation radiale | 2 voisins (lin√©aire) | 2 voisins (lin√©aire) |
| Interpolation angulaire | **Non** (plus proche voisin) | **Oui** (projections ¬±1 pond√©r√©es) |
| Nombre de voisins effectifs | 2 | **~4-6** (2 radial √ó 2-3 angulaire) || Continuit√© | Discontinuit√©s angulaires | **Liss√©e** |

```python
def recon_fourier_slice_bilinear(sinogram, angles, nbvox, voxsize, pixsize):
    """
    Reconstruction tomographique par th√©or√®me de Fourier Slice avec
    interpolation par lissage angulaire et interpolation lin√©aire radiale
    (approximation de l'interpolation bilin√©aire en coordonn√©es polaires).

    Cette version interpole les valeurs Fourier en utilisant les projections
    voisines (¬±1 angle) pond√©r√©es par la proximit√© angulaire, combin√©e √† une
    interpolation lin√©aire sur les fr√©quences radiales. Ce n'est pas une
    interpolation bilin√©aire au sens strict (4 voisins sur une grille
    r√©guli√®re), mais plut√¥t une approximation adapt√©e aux coordonn√©es polaires
    de l'espace de Fourier.
    """

    n_angles, n_pixels = sinogram.shape

    # √âTAPE 1 : Transform√©e de Fourier 1D de chaque projection
    fourier_projections = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(sinogram, axes=1), axis=1), axes=1)
    freq_rad = np.fft.fftshift(np.fft.fftfreq(n_pixels, d=pixsize))

    # √âTAPE 2 : Grille de Fourier 2D (coordonn√©es cart√©siennes)
    freq_2d = np.fft.fftshift(np.fft.fftfreq(nbvox, d=voxsize))
    kx, ky = np.meshgrid(freq_2d, freq_2d)

    # Conversion cart√©sien ‚Üí polaire
    k_radius = np.sqrt(kx**2 + ky**2)
    k_angle = np.arctan2(ky, kx)

    # √âTAPE 3 : Remplissage avec lissage angulaire + interpolation lin√©aire radiale
    fourier_2d = np.zeros((nbvox, nbvox), dtype=complex)

    for ia, angle_rad in enumerate(angles):
        angle_diff = np.abs(k_angle - angle_rad)
        angle_diff = np.minimum(angle_diff, 2*np.pi - angle_diff)

        angular_tolerance = np.pi / n_angles
        mask = angle_diff < angular_tolerance
        y_indices, x_indices = np.where(mask)

        for yi, xi in zip(y_indices, x_indices):
            k_r = k_radius[yi, xi]

            if k_r > freq_rad.max() or k_r < freq_rad.min():
                continue

            # Interpolation lin√©aire en fr√©quence radiale
            idx_float = np.interp(k_r, freq_rad, np.arange(n_pixels))
            idx_left = int(np.floor(idx_float))
            idx_right = min(idx_left + 1, n_pixels - 1)
            frac = idx_float - idx_left

            value_interp = (1 - frac) * fourier_projections[ia, idx_left] + \
                           frac * fourier_projections[ia, idx_right]

            # Lissage angulaire : moyenne pond√©r√©e avec projections voisines
            ia_prev = (ia - 1) % n_angles
            ia_next = (ia + 1) % n_angles

            value_prev = (1 - frac) * fourier_projections[ia_prev, idx_left] + \
                         frac * fourier_projections[ia_prev, idx_right]

            value_next = (1 - frac) * fourier_projections[ia_next, idx_left] + \
                         frac * fourier_projections[ia_next, idx_right]

            # Moyenne pond√©r√©e (approximation en coordonn√©es polaires)
            weight_main = 1.0
            weight_prev = max(0, (angular_tolerance - angle_diff[yi, xi]) / angular_tolerance)
            weight_next = weight_prev

            weight_sum = weight_main + weight_prev + weight_next

            value_bilinear = (weight_main * value_interp +
                              weight_prev * value_prev +
                              weight_next * value_next) / weight_sum

            fourier_2d[yi, xi] += value_bilinear

    # √âTAPE 4 : Transform√©e de Fourier inverse 2D
    image = np.fft.fftshift(np.fft.ifft2(np.fft.ifftshift(fourier_2d))).real

    return image

# Reconstruction avec lissage angulaire + interpolation lin√©aire radiale (grille 96√ó96)
print("BONUS : Interpolation avec lissage angulaire + interp. lin√©aire radiale")
t_start = time.time()
image_fourier_bilinear = recon_fourier_slice_bilinear(sinogram_password, angles, nbvox, voxsize, pixsize)
t_end = time.time()
temps_fourier_bilinear = t_end - t_start

print(f"Temps (lissage angulaire, 96x96) : {temps_fourier_bilinear:.2f}s")
print(f"  vs nearest (Q6) : {temps_fourier_bilinear / temps_fourier:.2f}x")
print(f"  vs grille 2x (Q8) : {temps_fourier_bilinear / temps_fourier_2x_q8:.2f}x")
```
