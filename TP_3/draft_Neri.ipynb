## TP3.1 Décomposition QR par la méthode de Householder

#a. Orthogonalité des matrices de réflexion $Q_i$

Dans cette question nous cherchons à démontrer l'orthogonalité des matrices de réflexion de la décomposition QR, autrement dit :

$$
Q_i^TQ_i=I
$$
\
Les matrices $Q_i$ sont par blocs diagonales,
\
\
$$
Q_i=\begin{bmatrix}
I_i & 0\\
0 & H_{m,i}
\end{bmatrix}
$$
\
Les transposées sont donc,
\
\
$$
Q_i^T=\begin{bmatrix}
I_i^T & 0\\
0 & H_{m,i}^T
\end{bmatrix}
$$
\
Le produit des matrices par blocs diagonales se fait bloc par bloc, nous avons donc,
\
\
$$
Q_i^TQ_i=\begin{bmatrix}
I_i^TI_i & 0\\
0 & H_{m,i}^TH_{m,i}
\end{bmatrix}
$$
\
Évidemment $I_i$ est orthogonale, $I_i^TI_i=I_i$, ainsi le problème se réduit à démontrer l'orthogonalité de $H_{m,i}$.
\
\
\
$$
H_{m,i}=I_{m-i}-2\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}
$$
\
Avec $\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}$, le projecteur du vecteur $\mathbf{v}_{m,i}$ de dimension $(m-i)\times1$
\
\
Par linéarité de la transposition,
\
\
$$
H_{m,i}^T=I_{m-i}^T-2\frac{(\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T)^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}
$$
\
Or&nbsp;&nbsp;$I_{m-i}^T=I_{m-i}$&nbsp;&nbsp;et&nbsp;&nbsp;$(\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T)^T=(\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T)$&nbsp;, donc on obtient,
\
\
$$
\begin{aligned}
H_{m,i}^T&= I_{m-i}-2\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\\
         &= H_{m,i}
\end{aligned}
$$
\
\
$H_{m,i}$ est symétrique, donc,
\
\
$$
\begin{aligned}
H_{m,i}^TH_{m,i}&= (H_{m,i})^2\\
                &= \left(I_{m-i}-2\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\right)^2\\
                &= I_{m-i}-4\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}+4\left(\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\right)^2
\end{aligned}
$$
\
Or $P=\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}$ est un projecteur, il vérifie $P^2=P$, donc,
\
\
$$
\begin{aligned}
H_{m,i}^TH_{m,i}&= I_{m-i}-4\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}+4\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\\
                &=I_{m-i}
\end{aligned}
$$
\
$H_{m,i}$ est donc orthogonale.
\
\
\
Finalement,
\
\
$$
\begin{aligned}
Q_i^TQ_i&= \begin{bmatrix}
        I_i & 0\\
        0 & I_{m-i}
      \end{bmatrix}\\
    &= I_m
\end{aligned}
$$
\
$Q_i$ est **orthogonale**.
