# TP3.1 Décomposition QR par la méthode de Householder

##a. Orthogonalité des matrices de réflexion $Q_i$

Dans cette question nous cherchons à démontrer l'orthogonalité des matrices de réflexion de la décomposition QR, autrement dit :

$$
Q_i^TQ_i=I
$$
\
Les matrices $Q_i$ sont par blocs diagonales,
\
\
$$
Q_i=\begin{bmatrix}
I_i & 0\\
0 & H_{m,i}
\end{bmatrix}
\tag{3.2}
$$
\
Les transposées sont donc,
\
\
$$
Q_i^T=\begin{bmatrix}
I_i^T & 0\\
0 & H_{m,i}^T
\end{bmatrix}
$$
\
Le produit des matrices par blocs diagonales se fait bloc par bloc, nous avons donc,
\
\
$$
Q_i^TQ_i=\begin{bmatrix}
I_i^TI_i & 0\\
0 & H_{m,i}^TH_{m,i}
\end{bmatrix}
$$
\
Évidemment $I_i$ est orthogonale, $I_i^TI_i=I_i$, ainsi le problème se réduit à démontrer l'orthogonalité de $H_{m,i}$.
\
\
\
$$
H_{m,i}=I_{m-i}-2\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}
\tag{3.3}
$$
\
Avec $\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}$, le projecteur du vecteur $\mathbf{v}_{m,i}$ de dimension $(m-i)\times1$
\
\
Par linéarité de la transposition,
\
\
$$
H_{m,i}^T=I_{m-i}^T-2\frac{(\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T)^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}
$$
\
Or&nbsp;&nbsp;$I_{m-i}^T=I_{m-i}$&nbsp;&nbsp;et&nbsp;&nbsp;$(\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T)^T=(\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T)$&nbsp;, donc on obtient,
\
\
$$
\begin{aligned}
H_{m,i}^T&= I_{m-i}-2\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\\
         &= H_{m,i}
\end{aligned}
$$
\
\
$H_{m,i}$ est symétrique, donc,
\
\
$$
\begin{aligned}
H_{m,i}^TH_{m,i}&= (H_{m,i})^2\\
                &= \left(I_{m-i}-2\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\right)^2\\
                &= I_{m-i}-4\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}+4\left(\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\right)^2
\end{aligned}
$$
\
Or $P=\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}$ est un projecteur, il vérifie $P^2=P$, donc,
\
\
$$
\begin{aligned}
H_{m,i}^TH_{m,i}&= I_{m-i}-4\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}+4\frac{\mathbf{v}_{m,i}\mathbf{v}_{m,i}^T}{\mathbf{v}_{m,i}^T\mathbf{v}_{m,i}}\\
                &=I_{m-i}
\end{aligned}
$$
\
$H_{m,i}$ est donc orthogonale.
\
\
\
Finalement,
\
\
$$
\begin{aligned}
Q_i^TQ_i&= \begin{bmatrix}
        I_i & 0\\
        0 & I_{m-i}
      \end{bmatrix}\\
    &= I_m
\end{aligned}
$$
\
$Q_i$ est **orthogonale**.



##b. Démonstration de l'équation $(3.5)$ et orthogonalité de $Q$.

Ici nous cherchons à démontrer l'expression de $Q$ suivante :
\
\
$$
\newcommand{\Q}{Q_0^TQ_1^T\cdots Q_{n-2}^TQ_{n-1}^T}
Q=\Q
\tag{3.5}
$$
\
\
$Q$ est la matrice telle que $A=QR$, et nous savons d'après l'énoncé que :  
\
$$
\newcommand{\QT}{Q_{n-1}Q_{n-2}\cdots Q_1Q_0}
\QT A=R
\tag{3.1}
$$
\
$$
\begin{aligned}
  (3.1) &\iff A=(\QT)^{-1}R\\
        &\iff Q=(\QT)^{-1}
\end{aligned}
$$
\
Les $Q_i$ sont orthogonales donc $Q_i^{-1}=Q_i^T$, ce qui donne pour $Q$ :
\
\
$$
\begin{aligned}
   Q&= (\QT)^{-1}\\
    &= Q_0^{-1}Q_1^{-1}\cdots Q_{n-2}^{-1}Q_{n-1}^{-1}\\
  Q&= \Q
\end{aligned}
$$
\
L'équation $(3.5)$ s'obtient à partir de l'équation $(3.1)$ et de l'orthogonalité des $Q_i$ démontrée à la question 3.1.a .
\
\
\
Voyons maintenant si $Q$ est orthogonale.
\
\
$$
\begin{aligned}
  Q^TQ&= (\Q)^T\Q \\
      &= Q_{n-1}Q_{n-2}\cdots Q_1\underbrace{Q_0Q_0^T}_{I_m}Q_1^T\cdots Q_{n-2}^TQ_{n-1}^T\\
      &= Q_{n-1}Q_{n-2}\cdots\underbrace{Q_1Q_1^T}_{I_m}\cdots Q_{n-2}^TQ_{n-1}^T\\
      &\vdots \\
  Q^TQ&=I_m
\end{aligned}
$$
\
$Q$ est orthogonale !

```python
def householder_qr(A):
    """
    Décomposition QR complète d'une matrice A par la méthode de Householder.

    Paramètres
    ----------
    A : matrice de dimension m x n (m >= n)

    Retourne
    --------
    Q : matrice orthogonale m x m
    R : matrice triangulaire supérieure m x n
    """
    A = A.astype(float)
    m, n = A.shape

    R = A.copy()
    Q = np.eye(m)

    for i in range(n):
        x = R[i:, i]
        sign = 1.0 if x[0] >= 0 else -1.0   #pour éviter les annulations catastrophiques
        v = sign * np.linalg.norm(x) * np.eye(len(x), 1).flatten() + x

        if v @ v < 1e-14:  # x est déjà un multiple de e1, rien à faire
            continue

        H = np.eye(len(x)) - 2 * np.outer(v, v) / (v @ v)
        R[i:, i:] = H @ R[i:, i:] #application qu'au sous espace de la colone i sans les i premier terme
        Q[i:, :]  = H @ Q[i:, :]  #Construction de Q^T

    return Q.T, R
```

##d.

```python
A = np.array([[1., 2., 3.],
             [4., 3., 5.],
             [6., 7., 8.],
             [9., 10., 11.]])
#---Méthode Householder---#
Qh, Rh = householder_qr(A)
print("Méthode Householder :")
print("Q=\n", np.round(Qh, 6))
```
