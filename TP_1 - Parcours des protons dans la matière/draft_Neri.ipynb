# 2. Homogénéité de $R_{CSDA}$

##2.1. Vérification des dimensions de $R_{CSDA}$

Si nous voulons vérifier la dimension de : $$R_{CSDA}= \int_{0}^{T_i}\frac{dT'}{\frac{S_{col}}{\rho}}$$

Il faut d'abbord vérifier la dimension du pouvoir d'arrêt collisionnel massique $\frac{S_{col}}{\rho}$ :

\begin{eqnarray}
\left[\frac{S_{col}}{\rho}\right]  &=& \left[-\left(\frac{dT}{\rho dx}\right)_{col}\right] \\
&=& \frac{E}{ML^{-3}L}  \\
&=& EL^2M^{-1}
\end{eqnarray}

Ce qui nous permet d'obtenir :

\begin{eqnarray}
\left[R_{CSDA}\right]&=&\left[\int_0^{T'}\frac{dT'}
{\frac{S_{col}}{\rho}}\right]\\
&=& \frac{\left[dT'\right]}{EL^2M^{-1}}\\
&=& \frac{E}{EL^2M^{-1}} \\
&=& ML^{-2}
\end{eqnarray}

Ansi la grandeur physique $R_{CSDA}$ semble être une masse surfacique.

##2.2. Discussion

Le résultat obtenu n'est pas une distance, ce à quoi nous nous attendions puisque $R_{CSDA}$ est nommée **_portée_** _par CSDA_ dans l'énoncé.

Nous remarquons que si la formule était exprimée selon le pouvoir d'arrêt collisionnel $S_{col}$ et non pas le pouvoir d'arrêt collisionnel **massique** $\frac{S_{col}}{\rho}$, nous obtiendrions bien un résultat homogène à une distance.

La formule de $R_{CSDA}$ nous permet en réalité d'obtenir une grandeur physique qui ne dépend pas de la densité massique du milieu dans lequel le faisceau de proton évolue.\

Mais l'information de la distance parcourue par les protons avant leur arrêt complet est bien comprise dans $R_{CSDA}$, l'intégrale est bornée entre une énergie $T'=0$, ce qui correspond au proton à l'arrêt, et une énergie initiale $T=T_i$; et si on divise ensuite ce résultat par la constante qu'est la masse volumique du milieu, on se retrouve donc bien avec la distance à laquelle le proton s'arrête complètement.
