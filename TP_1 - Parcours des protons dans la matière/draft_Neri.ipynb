# 2. Homogénéité de $R_{CSDA}$

##2.1. Vérification des dimensions de $R_{CSDA}$

Si nous voulons vérifier la dimension de : $$R_{CSDA}= \int_{0}^{T_i}\frac{dT'}{\frac{S_{col}}{\rho}}$$

Il faut d'abbord vérifier la dimension du pouvoir d'arrêt collisionnel massique $\frac{S_{col}}{\rho}$ :

\begin{eqnarray}
\left[\frac{S_{col}}{\rho}\right]  &=& \left[-\left(\frac{dT}{\rho dx}\right)_{col}\right] \\
&=& \frac{E}{ML^{-3}L}  \\
&=& EL^2M^{-1}
\end{eqnarray}

Ce qui nous permet d'obtenir :

\begin{eqnarray}
\left[R_{CSDA}\right]&=&\left[\int_0^{T_i}\frac{dT'}
{\frac{S_{col}}{\rho}}\right]\\
&=& \frac{\left[dT'\right]}{EL^2M^{-1}}\\
&=& \frac{E}{EL^2M^{-1}} \\
&=& ML^{-2}
\end{eqnarray}

Ansi la grandeur physique $R_{CSDA}$ semble être homogène à une masse surfacique.

##2.2. Discussion

Le résultat obtenu n'est pas une distance, ce à quoi nous nous attendions puisque $R_{CSDA}$ est nommée "**portée** par CSDA" dans l'énoncé.

Nous remarquons que si était utilisé dans la formule le pouvoir d'arrêt collisionnel $S_{col}$ et non pas le pouvoir d'arrêt collisionnel **massique** $\frac{S_{col}}{\rho}$, nous obtiendrions bien un résultat homogène à une distance.

$R_{CSDA}$ n'est donc pas une distance, mais plutôt la quantité de matière qu'il faut traverser par unité de surface pour qu'un proton initialement à l'énergie $T_i$ soit complètement arrêté. L'information de la distance parcourue par les protons avant leur arrêt complet est bien comprise dans $R_{CSDA}$, l'intégrale est bornée entre une énergie $T'=0$, ce qui correspond au proton à l'arrêt, et une énergie initiale $T'=T_i$. Et si on multiplie ensuite ce résultat par la constante qu'est la masse volumique du milieu, on se retrouve donc bien avec la distance à laquelle le proton s'arrête complètement.
\
\
Et il y a un intérêt à utiliser $R_{CSDA}$ tel que donné dans l'énoncé, et non pas comme une distance.\
D'après l'équation (4) de l'énoncé, $\frac{S_{col}}{\rho} \propto \frac{n_e}{\rho}$,\
or $n_e=\rho N_A\sum\limits_{i}\frac{w_iZ_i}{A_i}$, et donc : $$\frac{S_{col}}{\rho} \propto N_A\sum\limits_{i}\frac{w_iZ_i}{A_i}$$
\
Nous pouvons remarquer à partir de la première question que la contribution $N_A\sum\limits_{i}\frac{w_iZ_i}{A_i}$ ne varie que très peu entre l'eau et l'os, et de manière plus générale pour les milieux organiques.

La portée massique $R_{CSDA}$ dépend faiblement du matériau. Le pouvoir d'arrêt massique $\frac{S_{col}}{\rho}$ est proportionnel au rapport $\frac{Z}{A}$ qui varie peu entre les matériaux organiques. Il dépend aussi de l'énergie moyenne d'excitation $I$ du milieu, mais cette dépendance apparaît à l'intérieur d'un logarithme, ce qui en réduit fortement l'impact. Ainsi, pour une énergie donnée, $R_{CSDA}$ est proche d'une grandeur quasi universelle pour les tissus biologiques. En revanche, la portée classique $x=\frac{R_{CSDA}}{\rho}$ dépend directement de la masse volumique du milieu.







#3. Approximation du pouvoir d'arrêt total par le pouvoir d'arrêt collisionnel

##3.1 Pouvoir d'arrêt collisionnel et nucléaire
D'après les données obtenues grâce au PSTAR pour un rayonnement de proton de la gamme d'énergie 70-250 MeV dans de l'eau liquide, nous pouvons voir que le pouvoir d'arrêt électronique est environ $10^{4}$ fois plus important que le pouvoir d'arrêt nucléaire. Nous pouvons donc négliger le pouvoir d'arrêt nucléaire à côté du pouvoir d'arrête collisionnel pour la protonthérapie sur un corps humain (énergie de rayonnement 70-250 MeV et corps humain approché comme de l'eau liquide).
##3.2 Pouvoir d'arrêt radiatif
Mais il faut aussi se pencher sur le pouvoir d'arrêt radiatif, le _Bremsstrahlung_. C'est un phénomène observable pour des électrons à des énergie bien inférieures au MeV, et donc qui ne devrait pas pouvoir être négligé pour la gamme d'énergie que présentent les faisceaux de protonthérapie.  
En réalité cette perte d'énergie que représente le _Bremsstrahlung_ est du à l'accélération (ici décélération) de la particule, causée par des forces électro-statiques entre les particules du faisceaux et les particules de la cible. Et si ces forces sont à peu près les mêmes pour un proton et pour un électron, les masses quand à elles sont bien différentes, la masse d'un proton est ... fois supérieur à celle d'un électron et donc d'après le principe fondamental de la dynamique, la décélération est beaucoup moins importante pour un proton que pour un électron.
Et donc nous pouvons négliger le pouvoir d'arrêt radiatif pour un proton.
